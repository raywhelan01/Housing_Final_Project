```python
# Import ACS variables
var_url = "https://api.census.gov/data/2018/acs/acs5/profile/variables.json"
acs_vars = requests.get(var_url).json()
acs_vars = acs_vars["variables"]

vars_df = pd.DataFrame.from_dict(acs_vars, orient = 'index')
vars_df.index.name = 'code'
vars_df
```

Variable list:

- DP02_0001E = Number of households
- DP02_0015E = Average Household size
- DP04_0001E = Number of housing units


```python
# URL for ACS API call
var_list = ["DP02_0001E", "DP02_0015E", "DP03_0001E", "DP03_0010E", "DP03_0011E", "DP04_0001E", "DP04_0016E", "DP04_0050E"]


url_start = "https://api.census.gov/data/2018/acs/acs5/profile?get="
url_end =  "NAME&for=county:*&in=state:*&key=" + api_key

url = url_start

for var in var_list:
    url = url + var + ','

url = url + url_end

print(url)
```

```python
county65plus = censusdata.download('acs5', 2018, censusdata.censusgeo([('county', '*')]),
                                   ['DP05_0001E', 'DP05_0014PE', 'DP05_0015PE', 'DP05_0016PE', 'DP05_0021PE'],
                                   tabletype='profile')
county65plus.describe()
```
